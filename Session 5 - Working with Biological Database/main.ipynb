```python
from Bio import SeqIO

parseFile = SeqIO.parse("sequence.fasta", "fasta")
print(parseFile)

for n in parseFile:
    seq = n.seq

print(seq)
```

```python
readFile = SeqIO.read("sequence.fasta", "fasta")
print(readFile)
```

```python
seqA = readFile.seq
seqA_rna = seqA.transcribe()
print(seqA_rna)
```
