```python
from Bio import SeqIO
```

```python
seq0 = SeqIO.read("sequence.fasta", "fasta").seq
seqA = seq0[:20] + seq0[-4:]
seqB = seqA[:10] + seq0[round(len(seq0)/2):round(len(seq0)/2)+17]

print("Seq0: " + seq0)
print("SeqA: " + seqA)
print("SeqB: " + seqB)
```

```python
from Bio.SeqUtils import GC
print("AT: " + str(100-GC(seqA)))
```

```python
from Bio.Seq import Seq
from Bio import pairwise2
from Bio.pairwise2 import format_alignment

subseq = Seq("ATGCAAT")
alignment_seqA = pairwise2.align.localxx(seqA, subseq)

for n in alignment_seqA:
    print(format_alignment(*n))
```

```python
align_global = pairwise2.align.globalxx(seqA, seqB, score_only=True)
print(align_global)
```
